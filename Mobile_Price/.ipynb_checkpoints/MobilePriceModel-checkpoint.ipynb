# Imports and configs

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn.metrics import accuracy_score,classification_report
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

# Load data

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
#combine features
df_train['px_area'] = df_train.px_height * df_train.px_width
```

```python
df_test['px_area'] = df_train.px_height * df_train.px_width
```

# Features
1. ram
2. px_width,height = pixel area
3. battery_power

```python
df_train_feats = df_train[['ram','px_area','battery_power']]
df_train_feats
```

```python
X = df_train_feats
y = df_train.price_range
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size = .2,random_state = 42)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = LogisticRegression(max_iter = 1000)
model.fit(X_train_scaled,y_train)
y_pred = model.predict(X_test_scaled)
```

```python
print('Model Predictions: ',y_pred[:5])
```

```python
acc = accuracy_score(y_test,y_pred)
acc
```

```python
print(classification_report(y_test,y_pred))
```

```python
df_train['ram'].plot(kind = 'hist')
```

# Model on testing set

```python
X_train = df_train_feats
y_train = df_train.price_range
X_test = df_test[['ram','px_area','battery_power']]
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
test_model = LogisticRegression(max_iter = 1000)
test_model.fit(X_train_scaled,y_train)
```

```python
predictions = test_model.predict(X_test_scaled)
```

# Notes
1. Each of our feature, a random variable, have some sort of PDF (probability density function) thus there exists some distribution they fit. Though their function is not inheritly obvious there does exist some function that satisifes our random variable t
2. rarely do these distributions have textbook features they are almost always messy due to real world data
3. scaling the data is very important. we do this when some r.v data is widly different, numerically, compared to the other data. Scaling simply standardizes our random variable. this means we merely shift the mean, divide by its s.d, and changes the units. it does not make any variable 'norma' magically nor does it affect the data in any meaningful way such as removing crucial data like outliears or skewness.
